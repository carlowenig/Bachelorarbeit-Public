# Kohärente Zustände

```python
def plot_some_coherent_states():
    omegas = [100, 100, 100, 100, 10, 10]
    qs = [0, 0, 0.5, 0.5, 0, 0]
    ps = [0, 10, 0, 10, 0, 30]

    fig, axes = plt.subplots(3, 2, figsize=(8, 8), sharex=True, sharey=True)

    for ax, omega, q, p in zip(axes.flatten(), omegas, qs, ps):
        psi = create_coherent_psi(omega=omega, q=q, p=p)
        x = x_vals(1000).real
        psi_vals = psi(x)
        ax.plot(x, np.real(psi_vals), c="seagreen", label="Realteil")
        ax.plot(x, np.imag(psi_vals), c="dodgerblue", label="Imaginärteil")
        ax.text(0.05, .75, f"$\\omega = {omega}$\n$q = {q}$\n$p = {p}$", transform=ax.transAxes)

    axes[-1,0].legend(loc="lower left")
    
    for ax in axes[:,0]:
        ax.set_ylabel("$\\psi(x)$")

    for ax in axes[-1,:]:
        ax.set_xlabel("$x$")

    plt.subplots_adjust(hspace=0, wspace=0)
    plt.savefig("output/coherent_states.pdf", bbox_inches="tight")
    plt.show()

plot_some_coherent_states()
```

# Teste Normalisierungsfaktor $C$

```python
def analyze_normalization_factor():
    N_arr = np.logspace(1, 4, 100, dtype=int)
    C_vals = []
    dx_vals = []
    psi = create_coherent_psi() # Psi should be normalized

    for N in N_arr:
        x = x_vals(N)
        psi_vals = np.array([psi(x_val) for x_val in x])
        C = compute_C(psi_vals)
        C_vals.append(C)
        dx = x[1] - x[0]
        dx_vals.append(dx)

    C_vals = np.array(C_vals).real
    dx_vals = np.array(dx_vals).real
    sqrt_dx_vals = np.sqrt(dx_vals)

    C_err_vals = np.abs(C_vals - sqrt_dx_vals)

    plt.scatter(N_arr, C_err_vals, s=5, c="k", label="Datenpunkte")

    machine_precision = np.finfo(C_err_vals.dtype).eps
    plt.axhline(machine_precision, c="red", label=f"Maschinengenauigkeit $\\epsilon \\approx {fmt_tex(machine_precision)}$")

    N0 = N_arr[C_err_vals < machine_precision][0]
    plt.axvline(N0, c="green", label=f"$N_0 = {N0}$")

    plt.xlabel("$N$")
    plt.ylabel("$|C - \sqrt{\\Delta x}|$")
    plt.xscale("log")
    plt.yscale("log")
    plt.legend()

    plt.savefig("output/normierungsfaktor_konvergenz.pdf", bbox_inches="tight")
    plt.show()

analyze_normalization_factor()
```
